```python
newdf.loc[pd.isnull(newdf['NAME_TYPE_SUITE']),['NAME_TYPE_SUITE']] = 'Unaccompanied' #code for filling nan values in NMAE_TYPE_SUITE column
```

```python
round(100*(newdf.isnull().sum()/len(newdf.index)),2) #checking nan values in column in percentage
```

```python
#Removing rows with nan values as we have sufficient data to analyse.
#Hence rows with any null values can be removed as imputing them will introduce bias as nan percentage is 31.35
```

```python
newdf = newdf[newdf.isnull().sum(axis=1)<=0]
round(100*(newdf.isnull().sum()/len(newdf.index)),2)
```

```python
newdf.shape #checking the number of rows and columns left for analysis, which is sufficient
```

```python
q1 = newdf['AMT_INCOME_TOTAL'].quantile(0.25)
q3 = newdf['AMT_INCOME_TOTAL'].quantile(0.75)
iqr = q3-q1
low = q1-1.5*iqr
high = q3+1.5*iqr
newdf = newdf.loc[(newdf['AMT_INCOME_TOTAL']>low) & (newdf['AMT_INCOME_TOTAL']<high)]
```

```python
newdf.shape #checking count of rows & columns
```

```python
#Removing outliers
```

```python
q11 = newdf['DAYS_BIRTH'].quantile(0.25)
q31 = newdf['DAYS_BIRTH'].quantile(0.75)
iqr1 = q31-q11
low1 = q11-1.5*iqr1
high1 = q31+1.5*iqr1
newdf = newdf.loc[(newdf['DAYS_BIRTH']>low1) & (newdf['DAYS_BIRTH']<high1)]
```

```python
newdf.shape
```

```python
q12 = newdf['DAYS_EMPLOYED'].quantile(0.25)
q32 = newdf['DAYS_EMPLOYED'].quantile(0.75)
iqr2 = q32-q12
low2 = q12-1.5*iqr2
high2 = q32+1.5*iqr2
newdf = newdf.loc[(newdf['DAYS_EMPLOYED']>low2) & (newdf['DAYS_EMPLOYED']<high2)]
```

```python
newdf.shape
```

```python
newdf['DAYS_EMPLOYED'] = newdf['DAYS_EMPLOYED']*-2+newdf['DAYS_EMPLOYED'] #converting employed day to whole number
newdf['YEARS_EMPLOYED'] = round(newdf['DAYS_EMPLOYED']/365,0) #Rounding-off employed days to years
newdf['DAYS_BIRTH'] = newdf['DAYS_BIRTH']*-2+newdf['DAYS_BIRTH'] #converting DAYS_BIRTH to whole number
```

```python
bins = [0,3,7,15,50] #creating bins
group_names = ['fresher','experienced','lead','manager'] #naming bins
newdf['Grade'] = pd.cut(newdf['YEARS_EMPLOYED'], bins, labels=group_names) #appling bins
```

```python
from collections import Counter
x=Counter(newdf['TARGET']) 
print(x) #count of defaulters & non-defaluters
print(round(x[1]*100/x[0],2)) #Imbalance percentage
print(round(x[1]/x[0],2)) #Imbalance ration
```
